```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
```

```python
df['discounted_price'].head(5)
```

```python
df['discounted_price'].replace('₹', '', inplace=True, regex=True)
df['discounted_price'].replace(',', '', inplace=True, regex=True)
df['discounted_price']=df['discounted_price'].astype('float')
```

```python
df['discounted_price'].head(5)
```

```python
df['actual_price'].head(5)
```

```python
df['actual_price'].replace('₹', '', inplace=True, regex=True)
df['actual_price'].replace(',', '', inplace=True, regex=True)
df['actual_price']=df['actual_price'].astype('float')
```

```python
df['actual_price'].head(5)
```

```python
df['rating'].head(5)
```

```python
df['rating'].unique()
```

```python
df['rating'].replace(',', '.', inplace=True, regex=True)
df['rating'].replace('|', np.nan, inplace=True)
df['rating']=df['rating'].astype('float')
```

```python
df['rating'].head(5)
```

```python
df['rating_count'].head(5)
```

```python
df['rating_count'].replace(',', '', inplace=True, regex=True)
df['rating_count']=df['rating_count'].astype('float')
```

```python
df['rating_count'].head(5)
```

```python
df['discount_percentage'].head(5)
```

```python
df['discount_percentage'].replace('%', '', inplace=True, regex=True)
df['discount_percentage']=df['discount_percentage'].astype('float')
df['discount_percentage'].head(5)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(df['rating'].mean(), inplace=True)
df.fillna(df['rating_count'].mean(), inplace=True)
df.isnull().sum()
```

```python
df['product_name'].head(5)
```

```python
df['product_name']=df['product_name'].str.extract(r'^([A-Za-z -]+)',expand=True)
df['product_name'].head(5)
```

```python
df['category'].head(5)
```

```python
df['category']=df['category'].str.extract(r'^([A-Za-z -]+)',expand=True)
df['category'].head(5)
```

```python
category_counts = df['category'].value_counts().reset_index()
category_counts.columns = ['category', 'count']

# Plot using Plotly Express with a gradient color
fig = px.bar(
    category_counts.sort_values('count'),  # Optional: sort for better aesthetics
    x='count',
    y='category',
    orientation='h',
    color='count',
    color_continuous_scale='Viridis',  # Other options: 'Plasma', 'Cividis', 'Turbo', etc.
    title='Category Counts'
)

fig.update_layout(
    xaxis_title='Count',
    yaxis_title='Category',
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(255,255,255,1)',
    font=dict(size=11)
)

fig.show()
```

```python
rating_counts = df['rating'].value_counts().sort_index().reset_index()
rating_counts.columns = ['rating', 'count']

# Plot with Plotly
fig = px.bar(
    rating_counts,
    x='count',
    y='rating',
    orientation='h',
    color='count',
    color_continuous_scale='Blues',  # You can also use 'Viridis', 'Plasma', 'Inferno', etc.
    title='Rating Distribution'
)

fig.update_layout(
    xaxis_title='Count',
    yaxis_title='Rating',
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(255,255,255,1)',
    font=dict(size=14)
)

fig.show()
```

```python
fig = px.bar(
    df,
    x='rating',
    y='category',
    orientation='h',
    color='rating',
    color_continuous_scale='Viridis',  # Optional: choose a color scale
    title='Category Ratings'
)

fig.update_layout(
    xaxis_title='Rating',
    yaxis_title='Category',
    template='plotly_white',
    title_font_size=24,
    font=dict(size=14)
)
fig.update_yaxes(showgrid=True, gridcolor='lightgray')
fig.show()
```

```python
fig = px.bar(
    df,
    x='actual_price',
    y='category',
    orientation='h',
    color='actual_price',
    color_continuous_scale='Cividis',  # Professional, subtle color scale
    title='Actual Price by Category',
)

fig.update_layout(
    title_font=dict(size=24, family='Helvetica', color='darkblue'),
    xaxis_title='Actual Price',
    yaxis_title='Category',
    template='simple_white',
    font=dict(family='Arial', size=14, color='black'),
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(255,255,255,1)',
    margin=dict(l=100, r=40, t=60, b=40),
)

fig.update_xaxes(tickprefix="$", showgrid=True, gridcolor='lightgray')
fig.update_yaxes(showgrid=True, gridcolor='lightgray')

fig.show()
```

```python
fig = px.scatter(
    df,
    x='rating',
    y='actual_price',
    color='actual_price',  # Optional: color scale by price
    size_max=10,
    color_continuous_scale='Viridis',
    title='Actual Price vs. Rating',
)

fig.update_layout(
    title_font=dict(size=24, family='Helvetica', color='darkblue'),
    xaxis_title='Rating',
    yaxis_title='Actual Price',
    template='simple_white',
    font=dict(family='Arial', size=14, color='black'),
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(255,255,255,2)',
    margin=dict(l=60, r=40, t=60, b=60),
)


fig.update_yaxes(tickprefix="$", showgrid=True, gridcolor='lightgray')
fig.update_xaxes(showgrid=True, gridcolor='lightgray')

fig.show()
```

```python
data = df.groupby('category')['rating_count'].mean().reset_index()
fig = px.pie(data, names='category',values='rating_count',title='Rating Count per Category')
fig.show()
```

```python
discount_data = df.groupby('category')['discount_percentage'].mean().reset_index()

# Plot with Plotly Express as a pie chart
fig = px.pie(
    discount_data,
    names='category',
    values='discount_percentage',
    color='discount_percentage',

)

fig.update_traces(
    textinfo='percent+label',
    pull=[0.02] * len(discount_data),
    marker=dict(line=dict(color='#000000', width=1))
)

fig.update_layout(
    title_text='Average Discount Percentage per Category',
    title_font_size=20,
    showlegend=True,
    paper_bgcolor='white',
    plot_bgcolor='white',
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x='discounted_price',
    y='actual_price',
    z='discount_percentage',
    color='discount_percentage',  # Optional: color scale for clarity
    color_continuous_scale='Plasma',
    title='3D Scatter Plot: Discounted vs Actual Price vs Discount %',
)
fig.update_layout(
    title_font=dict(size=22, family='Helvetica', color='darkblue'),
    scene=dict(
        xaxis_title='Discounted Price',
        yaxis_title='Actual Price',
        zaxis_title='Discount Percentage',
        xaxis=dict(backgroundcolor="white", gridcolor="lightgray"),
        yaxis=dict(backgroundcolor="white", gridcolor="lightgray"),
        zaxis=dict(backgroundcolor="white", gridcolor="lightgray"),
    ),
    template='simple_white',
    font=dict(family='Arial', size=12, color='black'),
    margin=dict(l=0, r=0, t=50, b=0),
)

fig.show()
```
